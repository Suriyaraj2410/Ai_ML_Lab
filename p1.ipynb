```python
import numpy as np 
import matplotlib.pyplot as plt

MB=np.random.rand(300,80)
VD=np.random.normal(2,0.01,80).reshape(80,1)

Vector_D=np.sum(MB*VD.reshape(-1),axis=1)
Vector_mean=np.mean(Vector_D)
Vector_sd=np.std(Vector_D)

print(f'means = {Vector_mean}, standarddeviation = {Vector_sd}')

plt.figure(figsize=(10,10))
plt.hist(Vector_D,bins=5,edgecolor="green")

plt.xlabel("Value")
plt.ylabel("Frequency")

plt.title("Histogram of Vector D")

plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("./Data/Housing.csv")

print(data.describe())

price_grouped_data=data.groupby("furnishingstatus")["price"].mean()
sq_grouped_data=data.groupby("furnishingstatus")["area"].mean()
bedroom_grouped_data=data.groupby("furnishingstatus")["bedrooms"].mean()

print(bedroom_grouped_data)



def plotting(name):
    plt.figure(figsize=(10,10))
    sns.boxplot(data=data,x="furnishingstatus", y=name)
    plt.title(f"Graph of {name}")
    plt.show()

numberial_columns = data.columns[data.dtypes == "int64"]
print(numberial_columns)
plotting(numberial_columns[0])
plotting(numberial_columns[1])
plotting(numberial_columns[2])

sns.lmplot(x="price", y="area", hue="furnishingstatus",data=data)

plt.show()
    
```

```python
import pandas as pd


wine_red=pd.read_csv("./Data/winequality-red.csv",delimiter=";")
wine_white=pd.read_csv("./Data/winequality-white.csv",delimiter=";")

wine_red["type"]="red"
wine_white["type"]="white"

wine_red.head()



```

```python
wine_data=pd.concat([wine_red,wine_white],ignore_index=True)
categorical_value=wine_data.columns[wine_data.dtypes == "object"]

dummy_data=pd.get_dummies(wine_data["type"],dtype=int,drop_first=True)

preprocessed_data=pd.concat([wine_data,dummy_data],axis=1)

preprocessed_data.drop(["type"],axis=1,inplace=True)

preprocessed_data.tail()
```

```python

from sklearn.model_selection import train_test_split

# split the 'wine_df' dataframe into training and testing sets with radio 8:2 and sets the random seed for reproducibility
data_train, data_test = train_test_split(
    preprocessed_data, train_size=0.8, test_size=0.2, random_state=100
)

# shows the number of rows and columns in each set
data_train.shape, data_test.shape
```
